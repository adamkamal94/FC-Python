```python
#Extract Kepa's information, just like we did with the team name before
Keepa = data[data['short_name'] == 'Kepa'][['short_name','wage_eur','value_eur','player_positions','overall','age']]

#Assign Keepa's wage, position, rating and age to vairables
KeepaWage = Keepa['wage_eur'].item()
KeepaPos = Keepa['player_positions'].item()
KeepaRating = Keepa['overall'].item()
KeepaAge = Keepa['age'].item()
```

```python
#Create a list of goalkeepers, matching Kepa's position
longlist = data[data['player_positions'] == KeepaPos][['short_name','wage_eur','value_eur','player_positions','overall','potential','age']]

#Create a list of players that have a lower overall rank than Kepa
removals = longlist[longlist['overall'] <= KeepaRating].index
#Drop these players
longlist.drop(removals , inplace=True)

#Repeat above, but for players with a larger wage
removals = longlist[longlist['wage_eur'] > KeepaWage].index
longlist.drop(removals , inplace=True)

#Show me our potential replacements, sorted by lowest wages
longlist.sort_values('potential')
```

```python
def cheapReplacement(player, skillReduction = 0):
    
    #Get the replacee with the name provided in the argument
    replacee = data[data['short_name'] == player][['short_name','wage_eur','value_eur','player_positions','overall','potential','age']]

    #Assign the relevant details of this player to variables
    replaceePos = replacee['player_positions'].item()
    replaceeWage = replacee['wage_eur'].item()
    replaceeAge = replacee['age'].item()
    replaceeOverall = replacee['overall'].item() - skillReduction
    
    #Create the longlist of players that share the position
    longlist = data[data['player_positions'] == replaceePos][['short_name','wage_eur','value_eur','player_positions','overall','potential','age']]
    
    
    #Create list of players that do not meet the rating criteria and drop them from the longlist
    removals = longlist[longlist['overall'] <= replaceeOverall].index
    longlist.drop(removals , inplace=True)
    
    #Repeat for players with higher wages
    removals = longlist[longlist['wage_eur'] >= replaceeWage].index
    longlist.drop(removals , inplace=True)

    #Repeat for older players
    removals = longlist[longlist['age'] >= replaceeAge].index
    longlist.drop(removals , inplace=True)
    
    #Display the players that meet the requirements
    return longlist.sort_values('value_eur')
```
